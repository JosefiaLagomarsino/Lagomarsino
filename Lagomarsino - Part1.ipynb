#### <ins> Data visualisation:</ins>

1.- Interactive map of the Airnbnbs, filterd by only the ones with identified host; maximum rating; and price lower than the mean.

```python
import folium

#Filters:
df_v = df1[df1['host_identity_verified'] == 't']
max_r = df_v['review_scores_rating'].max()
df_v_max_r = df_v[df_v['review_scores_rating'] == max_r]
average_p = df_v_max_r['price'].mean()
df_final = df_v_max_r[df_v_max_r['price'] < average_p]


map_center = [df_final['latitude'].mean(), df_final['longitude'].mean()]
map_ = folium.Map(location=map_center, zoom_start=12)


for index, listing in df_final.iterrows():
    folium.Marker([listing['latitude'], listing['longitude']],
                  popup=f"{listing['name']} - {listing['neighbourhood_cleansed']} - Rating: {listing['review_scores_rating']} - Price: {listing['price']}",
                  icon=folium.Icon(color='blue')).add_to(map_)

map_
```

2.- Distribution of Airnbnb prices by neighborhood

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 4))
sns.barplot(x='neighbourhood_cleansed', y='price', data=df1, palette='pastel')
plt.gca().set_facecolor('black')
plt.grid(color='white', linestyle='--', linewidth=0.3)
plt.title('Distribution of Airnbnb prices by neighborhood')
plt.show()
```

It can be observed that the most expensive listings are situated in Fingal.

3.- Review scores frequency distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.histplot(df1['review_scores_rating'].dropna(), bins=30, color='pink')
plt.gca().set_facecolor('black')
plt.grid(color='white', linestyle='--', linewidth=0.3)
plt.title('Distribution of review Scores')
plt.xlabel('Review Scores')
plt.ylabel('Frequency')
plt.show()
```

It can be observed that the listings with highest scores are the ones with more reviews.

4.- Proportion of Superhost in the dataset

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(4, 4), facecolor='black')

counts = df1['host_is_superhost'].value_counts()
labels = {'t': 'Superhost', 'f': 'Not Superhost'}
plt.pie(counts, labels=counts.index.map(labels), autopct='%1.1f%%', colors=['lightgreen', 'pink'], startangle=90)
for text in plt.gca().texts:
    text.set_color('white')

plt.title('Proportion of Superhosts', color='white')
plt.show()
```

It can be observed the the mayority of the hosts are not SuperHost, representing an 87.1%.

5.- Distribution of rooms according to their types

```python
plt.figure(figsize=(8, 4))
df1['room_type'].value_counts().plot(kind='bar', color='pink')
plt.gca().set_facecolor('black')
plt.grid(color='white', linestyle='--', linewidth=0.3)
plt.title('Distribution of room per types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

It can be observed that the mayority of the listings are composed by entire home/ apartment and private rooms, and only a small proportion of shared and hotel rooms.
